```python
aqi_orig = pd.read_csv('annual_aqi_by_county_2020.csv')
aqi_orig
```

```python
aqi_orig.columns
```

```python
aqi_orig['% Good Days'] = aqi_orig['Good Days']/aqi_orig['Days with AQI']
aqi_orig['% Moderate Days'] = aqi_orig['Moderate Days']/aqi_orig['Days with AQI']
aqi_orig['% Unhealthy for Sensitive Groups Days'] = aqi_orig['Unhealthy for Sensitive Groups Days']/aqi_orig['Days with AQI']
aqi = aqi_orig.drop(['Year','Days with AQI', 'Good Days',
       'Moderate Days', 'Unhealthy for Sensitive Groups Days',
       'Unhealthy Days', 'Very Unhealthy Days', 'Hazardous Days', 'Days CO', 'Days NO2',
       'Days Ozone', 'Days PM2.5', 'Days PM10'],axis=1)
aqi
```

```python
state_encoding = pd.read_csv('state_encoding.csv').drop('abbrev',axis=1)
state_encoding
```

```python
# Note: 5 rows removed via inner join are Mexico and Virgin Island locations
aqi_new = aqi.merge(state_encoding,left_on='State',right_on='state')
aqi_new
```

```python
aqi_new = aqi_new.drop(['State','state'],axis=1)
aqi_new
```

```python
aqi_new['County_State'] = aqi_new['County'].str.strip() + ', ' + aqi_new['code']
aqi_new.drop(['County','code'],axis=1).to_csv('aqi_cleaned.csv',index=False)
```
