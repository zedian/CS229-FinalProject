```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
elec = pd.read_csv('countypres_2000-2020.csv')
elec
```

```python
elec = elec[elec['year']==2016]
elec = elec[(elec['party']=='DEMOCRAT') | (elec['party']=='REPUBLICAN')]
# elec[(elec['county_fips']>2000) & (elec['county_fips']<3000)]
```

```python
elec['% votes'] = elec['candidatevotes']/elec['totalvotes']
elec_new = pd.pivot_table(elec, values='% votes', index=['county_fips'],columns=['party'])
elec_new
```

```python
elec_new.index = elec_new.index.astype(int)
```

```python
elec_new.reset_index().to_csv('election_data.csv',index=False)
```
